# Las \ nos ayudan a organizarnos

```python
refran = "el perro de san roque no tiene rabo porque ramón rodriguez se lo ha robado, esto es un refrán típico en España."
```

Lineas de código muy largas dificultan su posterior legibilidad podemos cortar las cadenas con \

```python
refran = "el perro de san roque no tiene rabo "\
        "porque ramón rodriguez se lo ha robado,"\
        " esto es un refrán típico en España."
```

```python
print(refran)
```

Cuidado con añadir texto despues de la \ (incluso espacios)

```python
refran = "el perro de san roque no tiene rabo "\ 
        "porque ramón rodriguez se lo ha robado,"\
        " esto es un refrán típico en España."
```

El uso de \ nos permite "jugar" con la indentación pero es conveniente mantener la coherencia

Ejemplo de como NO hacerlo

```python
refran = "el perro de san roque no tiene rabo "\
  "porque ramón rodriguez se lo ha robado,"\
                 " esto es un refrán típico en España."
```

Pero si queremos añadir saltos de línea?

```python
refran = "el perro de san roque no tiene rabo \n"\
        "porque ramón rodriguez se lo ha robado,\n"\
        " esto es un refrán típico en España."
```

```python
print(refran)
```

También podemos usar '''

```python
refran = '''el perro de san roque no tiene rabo 
        porque ramón rodriguez se lo ha robado,
        esto es un refrán típico en España.'''
```

```python
print (refran)
```

pero teniendo cuidado con la indentación
