# Contextos

```python
variableGlobal = 7
```

```python
def contextoDefuncion(parametro):
  variable = 7
  print(f'la variable global que hemos definido antes valia {variableGlobal}')
  print(f'el valor de la variable local es {variable}')
  print(f'y el parámetro con el que hemos llamado a la función vale {parametro}')
```

```python
contextoDefuncion('entrada')
```

Se pueden modificar variables globales dentro de funciones usando el indicador **global**

```python
#ejemplo

def increment_variable():
  global variableGlobal # IMPORTANT !
  variableGlobal = variableGlobal + 3
  print(variableGlobal)
  
increment_variable()
```

## Return

```python
numero = sum([3,6])
print(numero)

matriz = [[1,0,1],[1,1,1],[1,1,1]]

def grafica_matriz(matriz):
  for elem_matriz in matriz:
    if type(elem_matriz) is list:
      grafica_matriz(elem_matriz)
      print("")
    else:
      print(f"[{elem_matriz}]", end="")

def suma_matrices(matriz):
  suma_acumulada = 0

  for elem_matriz in matriz:
    if type(elem_matriz) is list:
      suma_acumulada = suma_acumulada + suma_matrices(elem_matriz)
    else:
      suma_acumulada = suma_acumulada + elem_matriz
      
  return suma_acumulada
      
grafica_matriz(matriz)
suma_matrices(matriz)
```
