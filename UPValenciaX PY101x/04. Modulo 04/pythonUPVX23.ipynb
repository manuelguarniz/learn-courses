### Buscando el elemento más grande/pequeño de la lista

```python
lista = [9,13,2,7,124,-5]
print(max(lista))
print(min(lista))
```

como lo harías sin usar max/min?

```python
#TU RESPUESTA AQUÍ
lista = [9,13,2,7,124,-5]
num_max = 0
num_min = 0

for elem in lista:
  if (elem > num_max):
    num_max = elem
  if (elem < num_min):
    num_min = elem
    
print(num_max)
print(num_min)
```
